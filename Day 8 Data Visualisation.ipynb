```python
fig_4=plt.figure(figsize=(6,4))
axes_4=fig_4.add_axes([0,0,1,1])
#Creates a new figure (fig_4).
#The figsize=(6,4) sets the figure size to 6 inches wide and 4 inches tall.
#fig_4.add_axes([0, 0, 1, 1]):
#Adds a custom axis (axes_4) to the figure.
#[0, 0, 1, 1] defines the position and size of the axes:
#First two values (0, 0): The bottom-left corner of the axes (in figure coordinates, where (0,0) is bottom-left and (1,1) is top-right).
#Last two values (1, 1): The width and height of the axes (normalized from 0 to 1).
axes_4.set_xlabel('Temperature')
axes_4.set_ylabel('sales')
plt.title("temperature vs sales")
axes_4.plot(x_2,y_2)
```
